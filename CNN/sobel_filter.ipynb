```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
smile = cv2.imread('nz.jpg')
plt.imshow(smile)
```

```python
gray = cv2.cvtColor(smile, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_y = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
sobel_x
```

```python
sobel_y
```

```python
img_1 = cv2.filter2D(gray, -1, sobel_x)
img_2 = cv2.filter2D(gray, -1, sobel_y)
```

```python
fig, axes = plt.subplots(1,2,figsize=(20,5))
axes[0].set_title('horizontal')
axes[0].imshow(img_1, cmap = 'gray')
axes[1].set_title('vertical')
axes[1].imshow(img_2, cmap = 'gray')
plt.show()
```
